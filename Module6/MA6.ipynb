```python

import pandas as pd
import numpy as np

from scipy.sparse import lil_matrix

# Necessary import for linear regression model
from sklearn.linear_model import LinearRegression

# For splitting out training data
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
```

```python
#read in dataframe
who_df = pd.read_csv('../Module6/Life Expectancy Data.csv')
who_df
```

```python
#change column names
#make column names lowercase
old_names = who_df.columns
new_names = ['_'.join(c.split()).lower() for c in old_names]
who_df.columns = new_names
who_df.head()
```

```python
print(who_df.columns)
```

```python
who_df.info()
```

```python
#choose columns to use for analysis
who_df_filtered = who_df[['country', 'life_expectancy', 'income_composition_of_resources', 'schooling']]
who_df_filtered
```

```python
#average values for each
who_df_filtered = who_df_filtered.groupby(['country']).mean()
```

```python
#handle missing values
who_df_filtered = who_df_filtered.dropna()
```

```python
who_df_filtered
```

```python
who_df_filtered.describe()
```

```python
X = ['schooling','income_composition_of_resources']
y = 'life_expectancy'
```

```python
data_X = who_df_filtered[X]
data_y = who_df_filtered[y]
```

```python
#fit model
lm = LinearRegression()
lm.fit(data_X, data_y)
```

```python
#values for linear regression equation
print('intercept (b0) ', lm.intercept_)
coef_names = ['b1', 'b2']
print(pd.DataFrame({'Predictor': data_X.columns,
                   'coefficient Name': coef_names,
                   'coefficient Value': lm.coef_}))
```

```python
#train test split model
X_train, X_test, y_train, y_test = train_test_split(data_X, data_y, test_size=0.1, random_state=31337)
```

```python
#predicted life expectancy
predicted_le = lm.predict(X_test)
print("predicted values:", predicted_le)
```

```python
#actual vs predicted values
actual_pred = pd.DataFrame({'actual': y_test, 'predicted': predicted_le })
actual_pred
```

```python
#countries: Togo, Vietnam, Turkmenistan, Montenegro, Costa Rica
```

```python
sqr_errors = []

for country in actual_pred:
    error = actual_pred['predicted'] - actual_pred['actual']
    sqr_errors.append(error ** 2)

mse = np.mean(sqr_errors)
rmse = mse ** 0.5
print("Mean Square Error: %0.4f" % mse)
print("Root-Mean Square Error: %0.4f" % rmse)
```
